```python
from tkinter import * 
from tkinter.filedialog import askopenfilename
from keras.models import model_from_json
from PIL import Image, ImageTk
import numpy as np
import cv2
import joblib
from sklearn import *
import tkinter as tk

    
model1 = joblib.load('model_joblib_diabetes')
    
model = model_from_json(open('dr_model.json', 'r').read())
model.load_weights('dr_weights.hdf5' , by_name = True, skip_mismatch = True)

root = Tk()
root.title('Diabetic Prediction System')
root.geometry('1000x500')
root.configure(bg='gray')
message = StringVar()
message.set('Select an image and click Predict')

def openDiabeticPrediction():

    DiabeticPrediction = Toplevel(root)
    DiabeticPrediction.geometry("800x500")
    DiabeticPrediction.title("Diabetic Prediction")
    
    def show_entry_fields():
        p1=float(textbox1.get())
        p2=float(textbox2.get())
        p3=float(textbox3.get())
        p4=float(textbox4.get())
        p5=float(textbox5.get())
        p6=float(textbox6.get())
        p7=float(textbox7.get())
        p8=float(textbox8.get())
        p9=float(textbox9.get())
        p10=float(textbox10.get())
        p11=float(textbox11.get())
        
    
        result=model1.predict([[p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11]])
    
        if result == 0:
            label12 = tk.Label(DiabeticPrediction, text="Person is Healthy")
            label12.place(x=330, y=450)
        else:
            label12 = tk.Label(DiabeticPrediction, text="Person has Diabetic")
            label12.place(x=330, y=450)
            
    
    
    label0 = tk.Label(DiabeticPrediction, text = "Diabetes Prediction", bg = "black", fg = "white")
    label0.place(x=350, y=35)
    label1 = tk.Label(DiabeticPrediction, text="Enter the Age :")
    label1.place(x=200, y=70)
    label2 = tk.Label(DiabeticPrediction, text="Enter the Value of BMI :")
    label2.place(x=200, y=100)
    label3 = tk.Label(DiabeticPrediction, text="Enter the Value of Blood_Pressure(Systolic) :")
    label3.place(x=200, y=130)
    label4 = tk.Label(DiabeticPrediction, text="Enter the Value of Blood_Pressure(Diastolic) :")
    label4.place(x=200, y=160)
    label5 = tk.Label(DiabeticPrediction, text="Enter the Value of Fasting_Blood_Sugar :")
    label5.place(x=200, y=190)
    label6 = tk.Label(DiabeticPrediction, text="Enter the Value of Insulin_Dose(g) :")
    label6.place(x=200, y=220)
    label7 = tk.Label(DiabeticPrediction, text="Enter the Value of Cholesterol :")
    label7.place(x=200, y=250)
    label8 = tk.Label(DiabeticPrediction, text="Enter the Value of Kidny_Function(s Creatinine) :")
    label8.place(x=200, y=280)
    label9 = tk.Label(DiabeticPrediction, text="Enter the Value of Kidny_Function(eGFR) :")
    label9.place(x=200, y=310)
    label10 = tk.Label(DiabeticPrediction, text="Enter the Value of Liver_Function(ALT) :")
    label10.place(x=200, y=340)
    label11 = tk.Label(DiabeticPrediction, text="Pregnancy :")
    label11.place(x=200, y=370)

    textbox1 = tk.Entry(DiabeticPrediction)
    textbox1.place(x=500, y=70)
    textbox2 = tk.Entry(DiabeticPrediction)
    textbox2.place(x=500, y=100)
    textbox3 = tk.Entry(DiabeticPrediction)
    textbox3.place(x=500, y=130)
    textbox4 = tk.Entry(DiabeticPrediction)
    textbox4.place(x=500, y=160)
    textbox5 = tk.Entry(DiabeticPrediction)
    textbox5.place(x=500, y=190)
    textbox6 = tk.Entry(DiabeticPrediction)
    textbox6.place(x=500, y=220)
    textbox7 = tk.Entry(DiabeticPrediction)
    textbox7.place(x=500, y=250)
    textbox8 = tk.Entry(DiabeticPrediction)
    textbox8.place(x=500, y=280)
    textbox9 = tk.Entry(DiabeticPrediction)
    textbox9.place(x=500, y=310)
    textbox10 = tk.Entry(DiabeticPrediction)
    textbox10.place(x=500, y=340)
    textbox11 = tk.Entry(DiabeticPrediction)
    textbox11.place(x=500, y=370)


    btn1 = tk.Button(DiabeticPrediction, text='Predict', command=show_entry_fields)
    btn1.place(x=350, y=410)
    
    

def quit_app():
    root.quit()
    
def file_open():
    global filename
    filename = askopenfilename(title = "Select file", filetypes = (("png files","*.png"),("all files","*.*"),
                                                                  ("jpeg files","*.jpg")))
    
def show_img():
    dr_image = Image.open(filename)
    dr_image = dr_image.resize((300,300), Image.NEAREST)
    dr_image = ImageTk.PhotoImage(dr_image)
    img = Label(root, image=dr_image)
    img.image = dr_image
    img.place(x=50, y=150)
    
def crop_image_from_gray(img,tol=7):
    if img.ndim ==2:
        mask = img>tol
        return img[np.ix_(mask.any(1),mask.any(0))]
    elif img.ndim==3:
        gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        mask = gray_img>tol
        
        check_shape = img[:,:,0][np.ix_(mask.any(1),mask.any(0))].shape[0]
        if (check_shape == 0): # crop out because images too dark
            return img # return original image
        else:
            img1=img[:,:,0][np.ix_(mask.any(1),mask.any(0))]
            img2=img[:,:,1][np.ix_(mask.any(1),mask.any(0))]
            img3=img[:,:,2][np.ix_(mask.any(1),mask.any(0))]
    #         print(img1.shape,img2.shape,img3.shape)
            img = np.stack([img1,img2,img3],axis=-1)
    #         print(img.shape)
        return img
    
def prediction():
    dr_img = cv2.imread(filename)
    dr_img = crop_image_from_gray(dr_img,tol=7)
    dr_img = cv2.resize(dr_img,(224,224))
    dr_img = np.array(dr_img)/255
    dr_img = np.expand_dims(dr_img, axis=0)
    pred = model.predict(dr_img)
    ans = [round(i) for i in pred[0]]
    ans = sum(ans)
    if ans == 1:
        message.set('Person is Healthy')
    elif ans == 2:
        message.set('Person has Mild Diabetic Retinopathy')
    elif ans == 3:
        message.set('Person has Moderate Diabetic Retinopathy')
    elif ans == 4:
        message.set('Person has Severe Diabetic Retinopathy')
    elif ans == 5:
        message.set('Person has Proliferate Diabetic Retinopathy')
    
# Toolbar
toolbar = Frame(root, bd=1)
open_ = Image.open('open.png')
exit_ = Image.open('exit.jpg')

# Importing toolbar images
open_ = open_.resize((35,35), Image.NEAREST)
exit_ = exit_.resize((35,35), Image.NEAREST)
open_icon = ImageTk.PhotoImage(open_)
exit_icon = ImageTk.PhotoImage(exit_)

# Creating Toolbar Buttons
open_button = Button(toolbar, image=open_icon, command=file_open)
exit_button = Button(toolbar, image=exit_icon, command=quit_app)
open_button.image = open_button
exit_button.image = exit_button
open_button.grid(row=0, column=0, sticky='W')
exit_button.grid(row=0, column=1, sticky='W')
toolbar.grid(row=0, column = 0, sticky='W', ipadx=1000)

# Creating Show Image Button
show_img_button = Button(root, text='Show Image', bd=10, bg='light grey', height=2, width=10, command=show_img)
show_img_button.place(x=150, y=50)

# Creating Retinopathy Predict Button
show_img_button = Button(root, text='Predict Retinopathy', bd=10, bg='light grey', height=2, width=20, command=prediction)
show_img_button.place(x=700, y=50)

# Creating Diabetes Predict Button
show_img_button = Button(root, text='Predict Diabetes Here!!!', bd=10, bg='light grey', height=2, width=20, command=openDiabeticPrediction)
show_img_button.place(x=700, y=350)

# Creating label for showing Retinopathy prediction
prediction_label = Label(root, textvariable=message, bg='white', bd=14, height=7, width=40)
prediction_label.place(x=600, y=150)


root.mainloop()
```
